#### Importing necessary Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### Part One : Loading the data

```python
df1 = pd.read_csv("https://raw.githubusercontent.com/zalihat/cell_phone_dataset/master/cell_phones.csv")
```

```python
df1
```

```python
df2 = pd.read_csv("https://raw.githubusercontent.com/zalihat/cell_phone_dataset/master/population1.csv")
```

```python
df2
```

```python
df3 = pd.read_csv("https://raw.githubusercontent.com/open-numbers/ddf--gapminder--systema_globalis/master/ddf--entities--geo--country.csv")
```

```python
df3
```

## Part two: Joining the data

```python
# merging df1 and df2 
merged_df = pd.merge(df1, df2, how="inner", on=["Country", "year"])
```

```python
merged_df
```

```python
merged_df.head(10)
```

```python
merged_df.shape
```

```python
# renaming two columns in df3
renamed_df3 = df3.rename(columns={"country":"geo", "name": "Country"})
```

```python
renamed_df3
```

```python
# merging the two renamed columns in df3 to merged_df
merged_df1 = pd.merge(merged_df, renamed_df3[["geo", "Country"]], how="inner", on="Country")
```

```python
merged_df1
```

```python
merged_df1.shape
```

## Part three:  Making Features

```python
# calculate cell phone per person and add the column to the dataframe
merged_df1["cell_phones_per_person"] =(merged_df1["cell_phones_total"] / merged_df1["population"])
```

```python
merged_df1
```

```python
# modify the "geo" column with uppercase string
merged_df1["geo"] = merged_df1["geo"].str.upper()
```

```python
merged_df1
```
